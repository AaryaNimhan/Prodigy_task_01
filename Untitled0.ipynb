```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("world_population.csv", skiprows=4)

# List of countries you want
countries = [
    "China",
    "India",
    "United States",
    "Indonesia",
    "Pakistan",
    "Nigeria",
    "Brazil",
    "Russian Federation",
    "Mexico"
]

# Filter dataset
filtered = df[df["Country Name"].isin(countries)]

# Sort by 2020 population
filtered = filtered.sort_values(by="2020", ascending=False)

# Plot bar chart
plt.figure(figsize=(12,6))
plt.bar(filtered["Country Name"], filtered["2020"])

plt.title("Top Countries by Population in 2020")
plt.xlabel("Country")
plt.ylabel("Population")

plt.xticks(rotation=45)
plt.show()
```
